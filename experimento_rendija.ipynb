```python
import numpy as np
import matplotlib.pyplot as plt

def doble_rendija_bolitas(n_particulas, long_onda, dist_rendijas, dist_pantalla):
    '''
    x contiene en este caso 50 valores que tienen una distancia entre cada uno de 5 milimetros,
    cada valor que tiene x es una posible osicion en la pantalla 
    '''
    x = np.linspace(-0.5, 0.5, 50) 
    '''
    Probabilidad, calcula la pprobabilidad de interferencia de para toda posicion en la pantalla
    ''' 
    probabilidad = (np.cos((np.pi * dist_rendijas * x) / (long_onda * dist_pantalla)))**2
    '''
  Despues de con np.sum se normaliza sus valores para que estos sumen 1, los cuales serviran para la siguiente funcion, 
  para que muestre la istribucion de probabilidad en la pantalla 
    '''
    probabilidad /= np.sum(probabilidad)
    '''
   Impactos almacena almacena las posiciones que el parametro size indica,
   este parametro se utiliza para indicar que se lanzaron n_particulas, y por ultimo la funcion np.arndom.choice coloca la particula
   en una posicion aleatoria, basandose en la probabilidad
   ''' 
    impactos = np.random.choice(x, size=n_particulas, p=probabilidad)
    '''
    En esta parte de codigo se organiza la tabla para poder realizar la simulacion, en este saso,
    como utilizamos un laser rojo y no pusimos tanta distancia desde las rendijas, nos dio este resultado
    '''

    plt.figure(figsize=(10, 5))
    plt.scatter(impactos, np.random.rand(n_particulas) * 0.1, color='red', s=10, alpha=0.6)
    plt.title('Experimento de la doble rendija')
    plt.xlabel('Posicin en la pantalla (m)')
    plt.ylabel('Distribucin de impactos')
    plt.grid(True)


n_particulas = 1000
long_onda = 500e-9  
dist_rendijas = 1e-5  
dist_pantalla = 1  

doble_rendija_bolitas(n_particulas, long_onda, dist_rendijas, dist_pantalla)

```
